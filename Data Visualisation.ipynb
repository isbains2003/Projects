## Data Cleaning

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.info ()
```

```python
df_can.rename (columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)
```

```python
df_can.info ()
```

```python
df_can.columns = list (map (str, df_can.columns))
```

```python
df_can.set_index ('Country', inplace = True)
```

```python
df_can.head ()
```

```python
df_can['Total'] = df_can [df_can.columns [3:].tolist ()].sum (axis = 1)
```

```python
df_can.head ()
```

```python
years = df_can.columns [3:-1].tolist ()
```

```python
type (years [0])
```

```python
df_Continent = df_can.groupby ('Continent').sum ()
```

```python
df_Continent.head ()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','lightgreen', 'pink']
```

```python
df_Continent ['Total'].plot (kind = 'pie',figsize=(15, 6),
autopct='%1.1f%%', startangle=30, shadow=True, labels=None, pctdistance=1.22, colors = color_list)
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1)
plt.legend (labels = df_Continent.index.tolist (), loc = 'lower left')
plt.axis ('equal')
plt.show ()
```
